```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime 
import pandas_datareader as pdr
```

```python
start = datetime.datetime(2017,1,1)
end   = datetime.datetime(2022,6,6)
```

```python
meta=pdr.DataReader('META','yahoo',start,end)
amz=pdr.DataReader('AMZN','yahoo',start,end)
goog=pdr.DataReader('GOOGL','yahoo',start,end)
```

```python
goog.head()
```

```python
meta['High'].plot(label='Meta High price',figsize=(15,7))
amz['High'].plot(label='Amazon High price')
goog['High'].plot(label='Google High price')
plt.legend()
plt.show()
```

```python
meta['Volume'].plot(label='Meta Volume',figsize=(15,7))
amz['Volume'].plot(label='Amazon Volume')
goog['Volume'].plot(label='Google Volume')
plt.legend()
plt.show()
```

```python
meta['Total']=meta['Open']*meta['Volume']
amz['Total']=amz['Open']*amz['Volume']
goog['Total']=goog['Open']*goog['Volume']
```

```python
meta['Total'].plot(label='Meta Total', figsize=(15,7))
amz['Total'].plot(label='Amazon Total')
goog['Total'].plot(label='Google Total')
plt.legend()
plt.show()
```

```python
amz['Open'].plot(figsize=(15,7),label='Amazon Open price')
amz['MA50']=amz['Open'].rolling(50).mean()
amz['MA50'].plot(label='MA50')
plt.legend()
plt.show()
```
